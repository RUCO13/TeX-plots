```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import subprocess, os

sys.path.append('/media/labfiles/ruco/repos/TeX-plots/src')

#from splot import plots
x = np.arange(-10,10,.1)

yf = lambda x: np.sin(x)/(x)
y = np.array([yf(x[j]) for j in range(len(x))])

expdata = np.zeros((len(x),2))
expdata[:,0]=x
expdata[:,1]=y
np.savetxt("data/test-data.dat",expdata,delimiter=',')

#plot data
fig,ax = plt.subplots(1,1,figsize=(5,8))
# fig.set_facecolor('white')

ax.plot(x,y,'-ob',ms=7,mfc="w")
plt.show()
```

```python
currentdir=os.getcwd()
kv = {'data':currentdir+'/data/test-data.dat',
      'xlabel-user':'$x$',
      'ylabel-user':'$y$',
      'legen-user':'first plot',
      'color-user':'b'}

class plots:
    def __init__(self,variables):
        # for key,value in variables:
        #     self.key=value

        with open('../src/simple-plot.tex','r') as splot:
            text = splot.read()
            for key, value in variables.items():
                text = text.replace(key, value)
            with open('simple-plot-new.tex', 'w') as output:
                output.write(text)
    
        os.system("lualatex -shell-escape -file-line-error -output-directory=build-ruco %s"%("simple-plot-new.tex"))
        os.system("rm -R simple-plot-new.tex")

plots(kv)
```
